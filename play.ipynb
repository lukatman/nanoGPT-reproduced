```python
# dataset
with open("input.txt", "r") as f:
    text = f.read()
data = text[:1000]
print(data[:100])
```

```python
import tiktoken
enc = tiktoken.get_encoding('gpt2')
tokens = enc.encode(data)
print(tokens[:24])
```

Generate from tokens idx-shaped tensor (B, T), so it can be deployed for model training

```python
import torch
buf = torch.tensor(tokens[:24 + 1])
# x -> transformer input tensor, y -> target labels for each x
x = buf[:-1].view(4, 6) # just rearranges as 4x6 stack-ily
y = buf[1:].view(4,6)
print(x)
print(y)
```
